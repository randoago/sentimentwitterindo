```python
def sentences(teks):
    teks = teks.lower()
    tweet = re.sub(r'http\S+','', teks)#remove http
    tweet =re.sub('@[^\s]+',' ',tweet)#remove @
    tweet =re.sub('#[^\s]+',' ',tweet)#remove #
    tweet = re.sub(r'^rt[\s]+', '', tweet)#remove rt
    tweet = re.sub(r'\s\s+', ' ', tweet)#remove space
    tweet = re.sub('[^ a-zA-Z0-9]', ' ', tweet)
    #tweet = re.sub('+', ' ', tweet)#remove rt
    #removert = re.sub('rt',' ',removeat)
    #token=kata(tweet)
    return tweet
```
